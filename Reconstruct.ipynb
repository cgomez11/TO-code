```python
import glob
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.axes import Axes
from astropy.io import fits
from astropy import wcs
import warnings
warnings.filterwarnings("ignore")
ra_obj = 172.51287
dec_obj = 28.11336
worldcrd_obj = np.array([[ra_obj, dec_obj]]) 
main_path ='/home/oem/Documentos/Astronomy/CRTS/Data/NN/SN/'
#find folders that start with the corresponding number of each line 
lista_r1 = glob.glob(main_path + '*32')
lista_r1 = sorted(lista_r1)
lista_r2 = glob.glob(main_path + '*31')
lista_r2 = sorted(lista_r2)
lista_r3 = glob.glob(main_path + '*30')
lista_r3 = sorted(lista_r3)
list_of_rows = [lista_r1, lista_r2, lista_r3]
print('Top row:', lista_r1)
print('Middle row:', lista_r2)
print('Bottom row:', lista_r3)
```

```python
for index in lista_r2:
    print(index)
    list_ims = glob.glob(index+'/' + '*.fits')
    num_ims = len(list_ims)
    #fig = plt.figure(1)
    #define the size of the subplot given the number of images
    fig, axes = plt.subplots(1, num_ims)
    for k in range(num_ims):
        hdu_list = fits.open(list_ims[0])
        image_data = hdu_list[0].data
        w = wcs.WCS(hdu_list[0].header)
        ra_dec_crd = w.wcs.crval
        worldcrd_original_image = np.array([[ra_dec_crd[0] ,ra_dec_crd[1]]]) #shape = (1,2)
        pixcrd_original_image = w.wcs_world2pix(worldcrd_original_image, 1)
        print('Pixel crds original image centres RA/DEC:', pixcrd_original_image)
        #visualization
        inf_value = np.mean(image_data)- 2.5*np.std(image_data)
        sup_value = np.mean(image_data)+ 2.5*np.std(image_data)
        print('World crds OBJ:', worldcrd_obj)
        pixcrd_obj = w.wcs_world2pix(worldcrd_obj, 1)
        print('Pixel coordinates OBJ:',pixcrd_obj)
        x_obj = pixcrd_obj[0,0]
        y_obj = pixcrd_obj[0,1]
           
        axes[k].imshow(image_data, cmap='gray',origin='lower', vmin=inf_value, vmax = sup_value)
        axes[k].set_xticklabels([])
        axes[k].set_yticklabels([])
        #axes[k].scatter(x_obj, y_obj, facecolors='none', edgecolors='r', s=160)
        #axes.subplots_adjust(wspace=0, hspace=0)
    plt.show()
```

```python
plt.clf()
fig, axes = plt.subplots(3, 3, gridspec_kw = {'wspace':0, 'hspace':0}, figsize=(8,8))
for i, row in enumerate(list_of_rows):
    print('Row:', row)
    for j, col in enumerate(row):
        #print('Col:',col)
        list_ims = glob.glob(col+'/' + '*.fits')
        list_ims = sorted(list_ims) #visualize the same number of observation in the sequence
        #print(list_ims)
        myim = list_ims[0]
        print(myim)
        hdu_list = fits.open(myim)
        image_data = hdu_list[0].data
        inf_value = np.mean(image_data)- 2.5*np.std(image_data)
        sup_value = np.mean(image_data)+ 2.5*np.std(image_data)
        w = wcs.WCS(hdu_list[0].header)
        ra_dec_crd = w.wcs.crval
        worldcrd_original_image = np.array([[ra_dec_crd[0] ,ra_dec_crd[1]]]) #shape = (1,2)
        pixcrd_original_image = w.wcs_world2pix(worldcrd_original_image, 1)
        print('Grid position:',i,j, 'Pixel crds OI centres RA/DEC:', pixcrd_original_image)
        #visualization
        inf_value = np.mean(image_data)- 2.5*np.std(image_data)
        sup_value = np.mean(image_data)+ 2.5*np.std(image_data)
        #print('World crds OBJ:', worldcrd_obj)
        pixcrd_obj = w.wcs_world2pix(worldcrd_obj, 1)
        x_obj = pixcrd_obj[0,0]
        y_obj = pixcrd_obj[0,1]
        if i ==1 and j==1:
            axes[i,j].imshow(image_data, cmap='gray',origin='lower', vmin=inf_value, vmax = sup_value)
            axes[i,j].scatter(x_obj, y_obj, facecolors='none', edgecolors='r', s=160)
            axes[i,j].set_xticklabels([])
            axes[i,j].set_yticklabels([])
            print('Pixel coordinates OBJ:',pixcrd_obj)
        else: 
            axes[i,j].imshow(image_data, cmap='gray',origin='lower', vmin=inf_value, vmax = sup_value)
            #axes[i,j].set_aspect('equal')
            axes[i,j].set_xticklabels([])
            axes[i,j].set_yticklabels([])

#fig.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

```python
from imutils import build_montages
all_images = []
fig, axes = plt.subplots(3, 3, gridspec_kw = {'wspace':0, 'hspace':0}, figsize=(8,8))
for i, row in enumerate(list_of_rows):
    print('Row:', row)
    for j, col in enumerate(row):
        #print('Col:',col)
        list_ims = glob.glob(col+'/' + '*.fits')
        list_ims = sorted(list_ims) #visualize the same number of observation in the sequence
        #print(list_ims)
        myim = list_ims[0]
        print(myim)
        hdu_list = fits.open(myim)
        image_data = hdu_list[0].data
        all_images.append(image_data)
print(len(all_images))
```

```python
row_1 = np.hstack([all_images[0], all_images[1], all_images[2]])
row_2 = np.hstack([all_images[3], all_images[4], all_images[5]])
row_3 = np.hstack([all_images[6], all_images[7], all_images[8]])
collage = np.vstack([row_3, row_2, row_1])
inf_value = np.mean(collage)- 0.5*np.std(collage)
sup_value = np.mean(collage)+ 0.5*np.std(collage)
new_fig = plt.figure(figsize=(8,8))
plt.imshow(collage, cmap='gray',origin='lower', vmin=inf_value, vmax = sup_value)
plt.xticks([])
plt.yticks([])
plt.show()
```
